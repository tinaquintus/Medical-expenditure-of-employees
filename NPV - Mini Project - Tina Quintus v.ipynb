Q1. How many employees above 50 age smoke, what is their average expenditure on health management?

```python
print('The number of employees who smoke and are above the age 50 :')
df_age_50=data[(data['age']>50) & (data['smoke']=='yes')]['id'].count()
df_age_50
```

```python
print('Their average expenditure on health management :')
df_age_50=data[(data['age']>50) & (data['smoke']=='yes')]['expenditure'].mean()
df_age_50
```

# Inference:
The number of employees who smoke and are above the age 50 are 64 employees and their average 
expenditure on health management is Rs.38820.22
